```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import prettytable as pt
import math
```

```python
X0 = 0
U0 = 1
V0 = 0.5

def ftrue(x:float) -> float:
    return math.cos(math.sqrt(x))

def ftailor(x: float) -> float:
    return 1 - 0.5 * x + 1/24 * x**2 - 1/720 * x**3

def fpicar1(x:float) -> float:
    return 1 -0.5 * x

def fpicar2(x:float) -> float:
    return 1 - 0.5 * x

def fpicar3(x:float) -> float:
    return 1 - 0.5 * x + 1/16 * x**2


def fSet(x0:float, xmax : float, xstep : float, func) -> list[float]:
    return [func(x) for x in np.arange(x0, xmax+xstep, xstep)]

def feulerSet(x0:float, xmax : float, xstep : float) -> list[float]:
    x, u, v = X0, U0, V0
    result = [u]
    while x < xmax:
        newu = u + xstep * v
        result.append(newu)
        if x != 0:
            v += -(2 * v + u) / (4 * x) * xstep
        x += xstep
        u = newu
    x, u, v = X0, U0, V0
    while x > x0:
        newu = u - xstep * v
        result.insert(0, newu)
        if x!= 0:
            v -= -(2 * v + u) / (4 * x) * xstep
        x -= xstep
        u = newu

    return result


x0 = 0
xmax = 10
xstep = 0.01

xarr = np.arange(x0, xmax+xstep, xstep)
ftruearr = fSet(x0, xmax+xstep, xstep, ftrue)
ftailorarr = fSet(x0, xmax, xstep, ftailor)
fpicar1arr = fSet(x0, xmax+xstep, xstep, fpicar1)
fpicar2arr = fSet(x0, xmax+xstep, xstep, fpicar2)
fpicar3arr = fSet(x0, xmax+xstep, xstep, fpicar3)
fpicar4arr = fSet(x0, xmax+xstep, xstep, fpicar4)
eulerarr = feulerSet(x0, xmax, xstep)
fig = px.line()
fig.add_trace(go.Scatter(x=xarr, y=ftruearr, name="Точное решение"))
fig.add_trace(go.Scatter(x=xarr, y=eulerarr, name="Метод эйлера"))
fig.add_trace(go.Scatter(x=xarr, y=ftailorarr, name = "Формула Тейлора"))
fig.add_trace(go.Scatter(x=xarr, y=fpicar1arr, name="Метод Пикара 1-го порядка"))
fig.add_trace(go.Scatter(x=xarr, y=fpicar2arr, name="Метод Пикара 2-го порядка"))
fig.add_trace(go.Scatter(x=xarr, y=fpicar3arr, name="Метод Пикара 3-го порядка"))
fig.show()


```

```python
# Func:
# u' = x + u^3, u(0) = 0
X0, U0 = 0, 0

def fpicar1(x):
    return x ** 2 / 2

def fpicar2(x):
    return x ** 7 / 56
    + x ** 2 / 2

def fpicar3(x):
    return (x**(22))/(3863552)
    +(3*x**(17))/(106624)
    +(x**(12))/(896)
    +(x**7)/(56)
    +(x**2)/2

def fpicar4(x):
    return (x**(67))/(3863981943017739124736)
    +(9*x**(62))/(98677964914244452352)
    +(1081*x**(57))/(73440052228804050944)
    +(76623*x**(52))/(53388985337387155456)
    +(310503*x**(47))/(3244062457475366912)
    +(1069*x**(42))/(224099368435712)
    +(790667*x**(37))/(4145838316060672)
    +(11871*x**(32))/(1883187970048)
    +(361*x**(27))/(2101772288)
    +(47*x**(22))/(11941888)
    +(33*x**(17))/(426496)
    +(x**(12))/(896)
    +(x**7)/(56)
    +(x**2)/2

def fset(x0:float, xmax:float, xstep:float, func) -> list[float]:
    return [func(x) for x in np.arange(x0, xmax - xstep, xstep)]

def feulerSet(x0:float, xmax:float, xstep:float) -> list[float]:
    xarr = np.arange(0, xmax, xstep)
    u = U0
    result = [u]
    for x in xarr:
        u += xstep * (x + u) ** 3
        result.append(u)
    xarr = np.arange(0, x0 - xstep, -xstep)
    u = U0
    for x in xarr:
        u -= xstep * (x + u) ** 3
        result.insert(0, u)
    return result


x0 = 0
xstep = 1e-4
eps = 1e-4

eulerarr = []
x = 0
u = U0
u2 = U0
while (abs(u) < eps) or abs((u2 - u) / u) < eps:
    u2 += (2 * xstep) * (x + u2) ** 3
    u += xstep * (x + u) ** 3
    eulerarr.append(u)
    x += xstep
    u += xstep * (x + u) ** 3
    eulerarr.append(u)
    x += xstep

xmax = x
print(f"Max x: {xmax} with step: {xstep}")


xarr = np.arange(x0, xmax - xstep, xstep)

picar1arr = fset(x0, xmax, xstep, fpicar1)
picar2arr = fset(x0, xmax, xstep, fpicar2)
picar3arr = fset(x0, xmax, xstep, fpicar3)
picar4arr = fset(x0, xmax, xstep, fpicar4)

table = pt.PrettyTable()
table.field_names = ["x", "Euler", "Picar 1", "Picar 2", "Picar 3", "Picar 4"]
for i in range(len(xarr)):
    table.add_row(list(map(lambda x : f"{x:.10f}", [xarr[i], eulerarr[i], picar1arr[i], picar2arr[i], picar3arr[i], picar4arr[i]])))
print(table)
```

```python
X0 = 0
U0 = 0


eps = 1e-7
xtarget = 1.63
initStep = xtarget / 100


def findTarget(x0, u0, xstep, xtarget):
    x = x0
    u = u0
    uprev = u0
    try:
        while x < xtarget:
            uprev = u
            u += xstep * (x + u ** 3)
            x += xstep
    except Exception:
        return x, u
    if abs(x - xtarget) < abs(x - xstep - xtarget):
        return x, u
    return x - xstep, uprev

xstep = initStep
xprev, uprev = findTarget(X0, U0, xstep, xtarget)
print(f"Target reached with x: {xprev}, u: {uprev}, xstep: {xstep}")
xstep /= 2
while 1:
    x, u = findTarget(X0, U0, xstep, xtarget)
    print(f"Target reached with x: {x}, u: {u}, xstep: {xstep}")
    if abs(u - uprev) / uprev < eps:
        break
    xprev, uprev = x, u
    xstep /= 2
```
